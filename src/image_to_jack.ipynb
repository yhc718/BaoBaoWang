```python
from PIL import Image
img = Image.open('/image_in_binary_color.png')
new_img = img.resize((128, 64), Image.ANTIALIAS)
display(new_img)
```

```python
import cv2
import numpy as np

new_img = np.array(new_img)
image = cv2.cvtColor(new_img, cv2.COLOR_BGR2GRAY)
display(image)
```

```python
for i in range(63):
  for j in range(127):
    if (image[i, j] < 200):
      print(f"do Screen.drawRectangle({j * 4}, {i * 4}, {(j + 1) * 4}, {(i + 1) * 4});")
```
